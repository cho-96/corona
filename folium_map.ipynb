```python
import folium
import pandas as pd
contact =pd.read_csv('contacts_0720_wgs84.csv')
```

```python
import numpy as np
date_1 = np.sort(contact['date'].unique())
```

```python
date_list = []
heatmap_list = []
for date in np.sort(contact['date'].unique()):
    date_list.append(date)
    heatmap_list.append([[row['latitude'],row['longitude']] for index, row in contact[contact['date']==date].iterrows()])
    
```

```python
from folium import plugins

import geopandas as gpd
gu = gpd.read_file('Seoul_Gu25\Seoul_Gu25.shp')

map_seoul = folium.Map(location=[37.567241, 127.005659], zoom_start = 13) 
hm = plugins.HeatMapWithTime(heatmap_list,index=date_list,radius=10,max_opacity=0.8)
hm.add_to(map_seoul)

folium.GeoJson(
    gu,
    name='seoul_municipalities',
    style_function = lambda x: {'fillColor': '#00000000'}
).add_to(map_seoul)
map_seoul
```
