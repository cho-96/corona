```python
covid_tested = pd.read_csv('covid_tested.csv')
```

```python
covid_tested = covid_tested[['day','confirmed']]
```

```python
covid_tested = covid_tested[:100]
```

```python
last_seoul = pd.read_csv('proximity_Seoulcity_delay0_last.csv')
minus0_seoul = pd.read_csv('proximity_Seoulcity_delay0.csv')
minus1_seoul = pd.read_csv('proximity_Seoulcity_delay1.csv')
minus2_seoul = pd.read_csv('proximity_Seoulcity_delay2.csv')
minus3_seoul = pd.read_csv('proximity_Seoulcity_delay3.csv')
last_seoul_dong = pd.read_csv('proximity_Seoulcity_dong_delay0_last.csv')
minus0_seoul_dong = pd.read_csv('proximity_Seoulcity_dong_delay0.csv')
minus1_seoul_dong = pd.read_csv('proximity_Seoulcity_dong_delay1.csv')
minus2_seoul_dong = pd.read_csv('proximity_Seoulcity_dong_delay2.csv')
minus3_seoul_dong = pd.read_csv('proximity_Seoulcity_dong_delay3.csv')
last_seoul_gu = pd.read_csv('proximity_Seoulcity_gu_delay0_last.csv')
minus0_seoul_gu = pd.read_csv('proximity_Seoulcity_gu_delay0.csv')
minus1_seoul_gu = pd.read_csv('proximity_Seoulcity_gu_delay1.csv')
minus2_seoul_gu = pd.read_csv('proximity_Seoulcity_gu_delay2.csv')
minus3_seoul_gu = pd.read_csv('proximity_Seoulcity_gu_delay3.csv')
last_nation = pd.read_csv('proximity_nation_delay0_last.csv')
```

```python
len(last_nation)
```

```python
dataframe1 = pd.DataFrame([last_seoul,minus0_seoul,minus1_seoul,minus2_seoul,minus3_seoul,last_seoul_dong,minus0_seoul_dong,minus1_seoul_dong,minus2_seoul_dong,minus3_seoul_dong,last_seoul_gu,minus0_seoul_gu,minus1_seoul_gu,minus2_seoul_gu,minus3_seoul_gu,last_nation],
             ['last_seoul','minus0_seoul','minus1_seoul','minus2_seoul','minus3_seoul','last_seoul_dong','minus0_seoul_dong','minus1_seoul_dong','minus2_seoul_dong','minus3_seoul_dong','last_seoul_gu','minus0_seoul_gu','minus1_seoul_gu','minus2_seoul_gu','minus3_seoul_gu','last_nation'])
```

```python
dataframe1 = dataframe1.reset_index()
```

```python
dataframe1.columns = ['name','dataframe']
```

```python
dataframe1.head()
```

```python
import os
import matplotlib.pyplot as plt
for i in range(len(dataframe1)):
    if not os.path.isdir('average_distance_plot'):
        os.makedirs('average_distance_plot')
        fig, ax1 = plt.subplots(figsize=(15,7))
        plt.title(dataframe1.iloc[i,0]+' average distance')
        color = 'tab:red'
        ax1.set_xlabel('date')
        ax1.set_ylabel('daily confirmed case')
        ax1.bar(covid_tested['day'],covid_tested['confirmed'], color=color)
        ax2 = ax1.twinx()
        color = 'tab:blue'
        ax2.set_ylabel('average_distance')
        ax2.scatter(covid_tested['day'],dataframe1.iloc[i,1]['average_distance'], color=color)
        fig.tight_layout()
        figname = dataframe1.iloc[i,0]+'_average_distance.png'
        plt.savefig(os.path.join('average_distance_plot',figname))
    else:
        fig, ax1 = plt.subplots(figsize=(15,7))
        plt.title(dataframe1.iloc[i,0]+' average distance')
        color = 'tab:red'
        ax1.set_xlabel('date')
        ax1.set_ylabel('daily confirmed case')
        ax1.bar(covid_tested['day'],covid_tested['confirmed'], color=color)
        ax2 = ax1.twinx()
        color = 'tab:blue'
        ax2.set_ylabel('average_distance')
        ax2.scatter(covid_tested['day'],dataframe1.iloc[i,1]['average_distance'], color=color)
        fig.tight_layout()
        figname = dataframe1.iloc[i,0]+'_average_distance.png'
        plt.savefig(os.path.join('average_distance_plot',figname))
```
