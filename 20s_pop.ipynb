```python
def datahandle(columns, year, month):
    curDir = os.getcwd()
    zf = zipfile.ZipFile(curDir + '\\LOCAL_PEOPLE_' + str(year) + str(month) + '.zip')
    text_files = zf.infolist()
    list_ = []
    for text_file in text_files:
        print(text_file.filename)
        handled = pd.read_csv(zf.open(text_file.filename), encoding='cp949', usecols=columns).replace('*', 0)
        list_.append(handled)
    for listi in list_:
        listi.columns = ['기준일ID','시간대구분','행정동코드','집계구코드','total']
    handled = pd.concat(list_)
    columns.remove(0)
    columns.remove(1)
    columns.remove(2)
    columns.remove(3)
    handled.iloc[:,columns] = handled.iloc[:,columns].astype(float)
    handled.iloc[:,columns] = handled.iloc[:,columns].astype(int)
    handled['total'] = handled.iloc[:,columns].sum(axis=1)
    crunched = handled.iloc[:,[0,1,2,3]]
    crunched['total'] = handled['total']
    crunched.columns = ['기준일ID','시간대구분','행정동코드','집계구코드','total']
    crunched['기준일ID'] = crunched['기준일ID'].astype(str)
    crunched['date'] = [x[6:] for x in crunched['기준일ID']]      
    crunched.columns = ['wholedate','hour','행정동코드', '집계구코드','total','date']
    crunched['month'] = [x[4:6] for x in crunched['wholedate']]
    crunched['date'] = crunched['date'].astype(int)
    return crunched
```
