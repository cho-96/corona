```python
import pandas as pd
covid = pd.read_csv('COVID19_nation_0713.csv')
distance = pd.read_csv('distances_rev_d_175_rolling_upto14.csv')
```

```python
distance[['day1','dh_7day','dg_7day']][49:55]
```

```python
days = [1,26,27,50,51,81,82,90,100,111,112,120,130,137,138,165]
intervention_s = []
for i in range(1,15):
    intervention_seoul = covid[['day','date']]
    intervention_seoul['Seoul'] = covid['Seoul'].rolling(i,min_periods=1).mean()
    intervention_seoul = intervention_seoul.loc[intervention_seoul['day'].isin(days)]
    intervention_s.append(intervention_seoul)
break_seoul = intervention_s[0][['day','date']]
for i in range(14):
    newcolname = 'Seoul_'+str(i+1)
    print(newcolname)
    break_seoul[newcolname] = intervention_s[i]['Seoul']
for i in range(14):
    newcolname1 = 'dh_'+str(i+1)
    newcolname2 = 'dg_'+str(i+1)
    newcolname3 = 'dh_'+str(i+1)+'day'
    newcolname4 = 'dg_'+str(i+1)+'day'
    break_seoul[newcolname1] = distance[newcolname3]
    break_seoul[newcolname2] = distance[newcolname4]
```
